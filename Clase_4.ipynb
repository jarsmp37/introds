## Buscar registros duplicados

Por defecto, df.duplicated() considera una fila como duplicada si todos los valores de la fila coinciden exactamente con otra fila anterior en el DataFrame.

```python
#Con este comando se crea un booleano que muestra renglones duplicados
df.duplicated()
```

Muestra la suma de todas las filas duplicadas

```python
df.duplicated().sum()
```

Puedes especificar un subconjunto de columnas para considerar al determinar duplicados

```python
#Compara solo las columnas 'CustomerId', 'Surname' por lo que no 
#es requisito que sean las 18 columnas
#exactamente iguales, sólo las columnas 'CustomerId', 'Surname'
df.duplicated(subset=['CustomerId', 'Surname']).sum()
```

# Elimina los registros duplicados

```python
df_sin_dupl=df.drop_duplicates(subset=['CustomerId', 'Surname'])
df_sin_dupl
```

Se cuentan quitar duplicados, cabe aclarar que es sobre la original

```python
df_sin_dupl.duplicated().sum()
```

# Cambio del tipo de valores

A veces nuestros datos que observamos que son numéricos el df los identifica como caracter, se utiliza el comando info() para ver el Dtype

```python
df.info()
```

```python
df.head(3)
```

Se observa que columnas como HasCrCard es numérico, pero el dtype nos dice que es objeto, para verificar esto se utiliza el comando unique que nos muestra los valores únicos por columna

```python
df2=df[df["HasCrCard"]!='invalid_value']
df2
```

```python
df2['HasCrCard'].unique()
```

Se observa que los números están entre comillas, eso indica que son caracteres y no números.

Para cambiar el tipo se ocupa el comando astype( tipo de valor)

int  entero

float decimal

```python
df2['HasCrCard']=df2['HasCrCard'].astype(float)
```

```python
df2=df2.dropna(subset='HasCrCard')
df2
```

```python
df2['HasCrCard']=df2['HasCrCard'].astype(int)
```

si se verifica en unique de la columna tendremos que ya los considera como valores enteros, al no aparecer con comillas y el dtype object

```python
df2['HasCrCard'].unique()
```

Se visualiza que con info() ya cambio su tipo a int

```python
df2.info()
```

```python
df2["CreditScore"].unique()
```

```python
df3["CreditScore"].unique()
```

Muchas veces sale error un valor, es recomendable intentar pasar a valor float y después a int, para estos casos

Este proceso se recomienda hacerlo de forma manual, ya que el proceso de automatización puede generar que algunas columnas que si tienen caracteres las considere como valor entero, convirtiendo a hexadecimal el valor y modificando la estructura de la columna
