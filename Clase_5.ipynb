```python
#Se carga la base de datos

import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/jarsmp37/introds/main/Churn_Bank.csv')
df.head()
```

```python
#Eliminar una fila por el indexado

# Elimina filas con índices 0 y 1
df = df.drop([12203])
df
```

```python
# Restablecer el índice
df = df.reset_index(drop=True)
df
```

```python
#Eliminar columna en específico
df2= df.drop(columns=['RowNumber'])
df2
```

```python
#Renombrar una columna
df2 = df2.rename(columns={'Surname': 'Apellido'})
df2
```

```python
#Renombrar una columna
df2 = df2.rename(columns={"Geography":"Pais","Gender":"Género",	"Age":"Edad"})
df2
```

```python
# Crear un rango de fechas
start_date = '2024-08-01'
end_date = '2025-08-31'
date_range = pd.date_range(start=start_date, end=end_date, periods=12204)

# Convertir las fechas a formato string
date_range_str = date_range.strftime('%Y-%m-%d')
date_range_str 
```

```python
#Anexar una columna llamada date
df2['Fecha']=date_range_str 
df2.head()
```

```python
df2.info()
```

```python
df2['Fecha'] = pd.to_datetime(df2['Fecha'])
df2.info()
```

```python
df2
```

```python
# Filtrar por una fecha específica
df_filtrado = df2[df2['Fecha'] == '2024-12-01']
df_filtrado
```

```python
# Filtrar por un rango de fechas
df_rangofecha = df2[(df2['Fecha'] >= '2024-10-01') & (df2['Fecha'] <= '2025-03-30')]
df_rangofecha
```

```python
# Extraer el año
df2['año'] = df2['Fecha'].dt.year

# Extraer el mes
df2['mes'] = df2['Fecha'].dt.month

# Extraer el día
df2['dia'] = df2['Fecha'].dt.day

# Extraer el día de la semana
df2['díanombre'] = df2['Fecha'].dt.day_name()

df2.head()
```

```python
# Calcular la diferencia en días entre dos fechas
df2['days_difference'] = (df2['Fecha'] - pd.Timestamp('2024-01-01')).dt.days
df2.tail()
```

```python
df2
```

```python
# Diccionario para traducir los nombres de los días de la semana
traduccion = {
    'Monday': 'Lunes',
    'Tuesday': 'Martes',
    'Wednesday': 'Miércoles',
    'Thursday': 'Jueves',
    'Friday': 'Viernes',
    'Saturday': 'Sábado',
    'Sunday': 'Domingo',
    'France':'Francia',
    'Spain':'España',
    'Germany':'Alemania',
    'Female':'Mujer',
    'Male':'Hombre'
}
```

```python
df2["Género"].unique()
```

```python
# Reemplazar los nombres de los días en la columna 'dianombre'
df2['díanombre'] = df2['díanombre'].replace(traduccion)
df2
```

```python
df2['Género'] = df2['Género'].replace(traduccion)
df2['Pais'] = df2['Pais'].replace(traduccion)
df2
```

```python
#Guardar los resultados en un csv
df2.to_csv("Base_limpia.csv", index=False)
```
