```python
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/jarsmp37/introds/main/Churn_Bank.csv')
df
```

Limpiando los nan

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
df
```

Eliminando el texto invalid_value

```python
lista_col=df.columns
for i in lista_col:
    print(f"En la columna {i} los invalid_value son: {df[df[i] == 'invalid_value'].shape[0]}")
```

```python
for i in lista_col:
    df=df[df[i] != 'invalid_value']
df
```

Quitar los valores repetidos de la columna CustomerId

```python
df=df.drop_duplicates(subset=['CustomerId'])
df
```

Renombrando los nombres de las columnas al espaÃ±ol

```python
df.columns
```

estoy cambiando nombres

```python
df=df.rename(columns={'Surname': 'Apellido','Geography': 'Pais','Gender':'Genero','Age':'Edad',
                      'Tenure':'Estadia', 'NumOfProducts':'#Prod','HasCrCard':'TarjetaCredito',
                      'IsActiveMember':'Activo','EstimatedSalary':'Salario','Exited':'Salida',
                      'Complain':'Quejas','Satisfaction Score':'Nivel_satisf',
                      'Card Type':'Tipo_tarjeta','Point Earned':'Puntos'})
df
```

Quitamos la primera columna Row number

```python
df=df.drop(columns=['RowNumber'])
```

```python
df
```

ReindexaciÃ³n

```python
df = df.reset_index(drop=True)
df
```

Cambiando los tipos

```python
df.info()
```

```python
df.head(3)
```

```python
#Primero lo pasamos a float y despues a int en caso de dar error
df['CustomerId']=df['CustomerId'].astype(float)
df['CustomerId']=df['CustomerId'].astype(int)
df['CreditScore']=df['CreditScore'].astype(int)
df['Edad']=df['Edad'].astype(float)
df['Edad']=df['Edad'].astype(int)
df['Estadia']=df['Estadia'].astype(int)
df['Balance']=df['Balance'].astype(float)
df['Edad']=df['Edad'].astype(float)
df['Edad']=df['Edad'].astype(int)
```

```python
df.columns
```

```python
#Lista de columnas a int
int_col=['CustomerId', 'CreditScore', 'Edad',
       'Estadia', '#Prod', 'TarjetaCredito', 'Activo',
       'Salida', 'Quejas', 'Nivel_satisf', 'Puntos']

for i in int_col:
    df[i]=df[i].astype(float)
    df[i]=df[i].astype(int)
```

```python
df['Salario']=df['Salario'].astype(float)
```
