plt.figure(figsize=(8, 6)): Define el tamaño del gráfico (ancho y alto) en pulgadas.

data: Los datos que serán usados para crear el histograma.

# Boxplot (Diagrama de Caja)

Un **boxplot** (o diagrama de caja) es una herramienta de visualización que resume un conjunto de datos mostrando su distribución a través de cinco números clave:

1. **Mínimo**: El valor más pequeño (excluyendo valores atípicos).
2. **Primer cuartil (Q1)**: El 25% de los datos están por debajo de este valor.
3. **Mediana (Q2)**: El valor que separa la mitad superior de la mitad inferior de los datos.
4. **Tercer cuartil (Q3)**: El 75% de los datos están por debajo de este valor.
5. **Máximo**: El valor más grande (excluyendo valores atípicos).

Además, un **boxplot** también puede mostrar los valores atípicos (outliers), que son puntos de datos que están considerablemente por encima o por debajo del rango intercuartil (IQR).

## Componentes del Boxplot:
- **Caja**: Representa el rango entre el primer cuartil (Q1) y el tercer cuartil (Q3).
- **Línea dentro de la caja**: Representa la mediana (Q2).
- **Bigotes**: Muestran la variabilidad de los datos fuera de los cuartiles, extendiéndose hasta 1.5 veces el IQR.
- **Puntos fuera de los bigotes**: Son valores atípicos.

### Ejemplo de un Boxplot
El siguiente código crea un boxplot con datos generados aleatoriamente:


```python
# Crear el boxplot
plt.figure(figsize=(3, 6))
plt.boxplot(df['Edad'], patch_artist=True, notch=True, boxprops=dict(facecolor='lightblue', color='blue'))

# Personalizar el gráfico
plt.title('Boxplot de edad', fontsize=16)
plt.ylabel('Valores', fontsize=12)

# Mostrar el gráfico
plt.show()
```

plt.boxplot(): Crea el gráfico de boxplot.

patch_artist=True: Rellena la caja con color.

notch=True: Añade una muesca que muestra una aproximación visual del intervalo de confianza para la mediana.

boxprops: Personaliza el color y la apariencia de la caja.

```python
# Crear el boxplot
plt.figure(figsize=(8, 3))
plt.boxplot(df['Edad'], patch_artist=True, notch=True, boxprops=dict(facecolor='lightblue', color='blue'), vert=False)

# Personalizar el gráfico
plt.title('Boxplot de edad', fontsize=16)
plt.ylabel('Valores', fontsize=12)

# Mostrar el gráfico
plt.show()
```
