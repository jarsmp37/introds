# Gráfico de Barras (Bar Plot) en Matplotlib

Un gráfico de barras se utiliza para visualizar distribuciones de variables categóricas. A continuación se describe cómo crear un gráfico de barras y las diferentes características que puedes modificar en Matplotlib.

## Características Modificables:

1. **Datos de entrada**: La variable categórica que se representará.
2. **Tamaño de la figura**: Controla el tamaño del gráfico.
3. **Colores de las barras**: Se puede especificar el color de las barras individualmente o de forma global.
4. **Estilo de las barras**: Puedes personalizar el borde de las barras.
5. **Rotación de etiquetas**: Para hacer más legible el gráfico si las etiquetas son largas.
6. **Etiquetas del gráfico**: Títulos, etiquetas de los ejes y leyendas.
7. **Ancho de las barras**: Controla qué tan anchas o delgadas son las barras.
8. **Rejilla**: Añade o elimina la rejilla en el gráfico.


### Desglose de Parámetros de `plt.grid()`

- **`True`**:
  Indica que se debe mostrar la rejilla. Si se establece en `False`, la rejilla no se mostrará.

- **`which='both'`**:
  Especifica qué líneas de rejilla se deben mostrar. Puede tomar los siguientes valores:
  - `'major'`: Solo líneas de rejilla para marcas principales.
  - `'minor'`: Solo líneas de rejilla para marcas secundarias.
  - `'both'`: Muestra ambas, las líneas de rejilla para marcas principales y secundarias.

- **`axis='y'`**:
  Define qué eje se verá afectado por la rejilla. Puede ser:
  - `'x'`: Rejilla solo en el eje X.
  - `'y'`: Rejilla solo en el eje Y.
  - `'both'`: Rejilla en ambos ejes.

- **`linestyle='--'`**:
  Especifica el estilo de las líneas de la rejilla. Los estilos pueden ser:
  - `'-'`: Línea sólida (por defecto).
  - `'--'`: Línea discontinua.
  - `':'`: Línea punteada.
  - `'-.'`: Línea de punto y raya.
  
  En este caso, se utiliza una línea discontinua.

- **`linewidth=0.7`**:
  Controla el grosor de las líneas de la rejilla. Un valor más alto hace que las líneas sean más gruesas, mientras que un valor más bajo las hace más delgadas. El valor por defecto es `1.0`.

- **`alpha=0.7`**:
  Define la opacidad de las líneas de la rejilla. El rango va de `0` (totalmente transparente) a `1` (totalmente opaco). Un valor de `0.7` significa que las líneas serán semi-transparentes.


### Desglose de Parámetros de `plt.xticks()`

- **`rotation=45`**:
  Este parámetro especifica el ángulo de rotación de las etiquetas del eje X. Un valor de `45` grados hace que las etiquetas se inclinen, lo que puede mejorar la legibilidad cuando hay muchas etiquetas o si son largas.

- **`ha='right'`**:
  Este parámetro define la alineación horizontal de las etiquetas. Puede tomar los siguientes valores:
  - `'center'`: Alinea las etiquetas en el centro.
  - `'right'`: Alinea las etiquetas a la derecha.
  - `'left'`: Alinea las etiquetas a la izquierda.
  
  En este caso, se establece en `'right'`, lo que significa que las etiquetas estarán alineadas a la derecha, lo que puede ayudar a evitar solapamientos con otras etiquetas o elementos del gráfico.


```python
plt.figure(figsize=(15, 6)) 
df['Edad'].value_counts().plot(kind='bar', color='lightblue')
plt.title('Gráfico de Barras')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')


# Rotar las etiquetas del eje X si es necesario
plt.xticks(rotation=45, ha='right')


# Añadir rejilla opcional
plt.grid(True, which='both', axis='y', linestyle='--', linewidth=0.7, alpha=0.7)

# Mostrar el gráfico
plt.show()
```

```python
plt.figure(figsize=(15, 6))

# Ordenar las edades antes de crear el gráfico de barras
df['Edad'].value_counts().sort_index().plot(kind='bar', color='lightblue')

# Personalización del gráfico
plt.title('Gráfico de Barras')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')

# Rotar las etiquetas del eje X
plt.xticks(rotation=45, ha='right')

# Añadir rejilla
plt.grid(True, which='both', axis='y', linestyle='--', linewidth=0.7, alpha=0.7)

# Mostrar el gráfico
plt.show()
```

```python
plt.figure(figsize=(15, 6))

# Obtener las 10 edades más comunes y ordenarlas
df['Edad'].value_counts().nlargest(10).plot(kind='bar', color='lightblue')

# Personalización del gráfico
plt.title('Top 10 Edades más Frecuentes', fontsize=16)
plt.xlabel('Edad', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)

# Rotar las etiquetas del eje X
plt.xticks(rotation=45, ha='right')

# Añadir rejilla
plt.grid(True, which='both', axis='y', linestyle='--', linewidth=0.7, alpha=0.7)

# Mostrar el gráfico
plt.show()
```

```python

# Obtener las 10 edades más comunes y ordenarlas
df['Edad'].value_counts().nlargest(10).plot(kind='barh', color='lightblue')

# Personalización del gráfico
plt.title('Top 10 Edades más Frecuentes', fontsize=16)
plt.xlabel('Edad', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)

# Rotar las etiquetas del eje X
plt.xticks(rotation=45, ha='right')

# Añadir rejilla
plt.grid(True, which='both', axis='y', linestyle='--', linewidth=0.7, alpha=0.7)

# Mostrar el gráfico
plt.show()
```

```python
plt.figure(figsize=(15, 6))

# Obtener las 10 edades más comunes (ordenadas por frecuencia de mayor a menor)
df['Edad'].value_counts().nlargest(10).plot(kind='barh', color='lightblue')

# Personalización del gráfico
plt.title('Top 10 Edades más Frecuentes', fontsize=16)
plt.xlabel('Frecuencia', fontsize=12)
plt.ylabel('Edad', fontsize=12)

# Añadir rejilla
plt.grid(True, which='both', axis='x', linestyle='--', linewidth=0.7, alpha=0.7)

# Invertir el eje Y para que la barra más alta esté en la parte superior
plt.gca().invert_yaxis()

# Mostrar el gráfico
plt.show()
```

```python

# Obtener las 10 edades más comunes
top_10_edades = df['Edad'].value_counts().nlargest(10)

# Definir una lista de colores para las barras
colors = plt.cm.viridis(np.linspace(0, 1, len(top_10_edades)))

# Crear el gráfico de barras horizontal con colores múltiples
top_10_edades.plot(kind='barh', color=colors)

# Personalización del gráfico
plt.title('Top 10 Edades más Frecuentes', fontsize=16)
plt.xlabel('Frecuencia', fontsize=12)
plt.ylabel('Edad', fontsize=12)

# Añadir rejilla
plt.grid(True, which='both', axis='x', linestyle='--', linewidth=0.7, alpha=0.7)

# Invertir el eje Y para que la barra más alta esté en la parte superior
plt.gca().invert_yaxis()

# Mostrar el gráfico
plt.show()
```

```python

# Obtener las 10 edades más comunes
top_10_edades = df['Edad'].value_counts().nlargest(10)

# Definir una lista de colores para las barras
colors = plt.cm.inferno(np.linspace(0, 1, len(top_10_edades)))

# Crear el gráfico de barras horizontal con colores múltiples
ax = top_10_edades.plot(kind='barh', color=colors)

# Añadir etiquetas al final de cada barra
for index, value in enumerate(top_10_edades):
    ax.text(value, index, str(value), va='center')  # va='center' alinea verticalmente

# Personalización del gráfico
plt.title('Top 10 Edades más Frecuentes', fontsize=16)
plt.xlabel('Frecuencia', fontsize=12)
plt.ylabel('Edad', fontsize=12)

# Añadir rejilla
plt.grid(True, which='both', axis='x', linestyle='--', linewidth=0.7, alpha=0.7)

# Invertir el eje Y para que la barra más alta esté en la parte superior
plt.gca().invert_yaxis()

# Mostrar el gráfico
plt.show()
```

### Desglose del Comando `ax.text()`

- **`ax.text(value, index, str(value), va='center')`**:
  - **`value`**: Posición en el eje X donde se coloca la etiqueta (frecuencia).
  - **`index`**: Posición en el eje Y (la edad).
  - **`str(value)`**: Convierte el valor a cadena para mostrarlo como texto.
  - **`va='center'`**: Alinea verticalmente el texto al centro de la barra.
