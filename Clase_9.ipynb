## Gráfico de distribución (KDE Plot)
Este gráfico muestra una estimación de la densidad de una variable continua, lo que ayuda a visualizar la distribución de los datos de manera más suave.

```python
# Crear figura y subplots (2 filas, 2 columnas)
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# KDE para Balance
sns.kdeplot(df['Balance'], fill=True, ax=axes[0, 0])
axes[0, 0].set_title('Distribución de Balance')

# KDE para Salario
sns.kdeplot(df['Salario'], fill=True, ax=axes[0, 1])
axes[0, 1].set_title('Distribución de Salario')

# KDE para CreditScore
sns.kdeplot(df['CreditScore'], fill=True, ax=axes[1, 0])
axes[1, 0].set_title('Distribución de CreditScore')

# KDE para Edad
sns.kdeplot(df['Edad'], fill=True, ax=axes[1, 1])
axes[1, 1].set_title('Distribución de Edad')

# Ajustar el layout para evitar superposición de títulos y etiquetas
plt.tight_layout()

# Mostrar gráfico
plt.show()
```

## Gráficos de Pastel (Pie Charts)

Los gráficos de pastel son útiles para visualizar la distribución de datos categóricos en proporciones. Cada "porción" del pastel representa una categoría, y el tamaño de la porción es proporcional a la frecuencia o cantidad de esa categoría.


```python
# Contar la cantidad de clientes por género
gender_counts = df['Genero'].value_counts()

# Gráfico de pastel
plt.figure(figsize=(6,6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90,
         colors=['skyblue', 'lightgreen'])
plt.title('Distribución por Género')
plt.show()
```

```python
# Contar la cantidad de clientes por país
pais_counts = df['Pais'].value_counts()

# Gráfico de pastel
plt.figure(figsize=(6,6))
plt.pie(pais_counts, labels=pais_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'skyblue', 'lightgreen', 'yellow'])
plt.title('Distribución por País')
plt.show()
```

```python
# Contar la cantidad de clientes por tipo de tarjeta
card_type_counts = df['Tipo_tarjeta'].value_counts()

# Gráfico de pastel
plt.figure(figsize=(5,5))
plt.pie(card_type_counts, labels=card_type_counts.index, autopct='%1.1f%%', startangle=90, colors=['purple', 'orange', 'green', 'blue'])
plt.title('Distribución por Tipo de Tarjeta')
plt.show()
```

## Gráficos de Violín (Violin Plots)

El gráfico de violín es una combinación entre un boxplot y un gráfico de densidad kernel (KDE). Permite visualizar la distribución de una variable continua a través de varias categorías, mostrando tanto la forma de la distribución como su dispersión y densidad. A diferencia de un boxplot tradicional, el gráfico de violín muestra una mayor cantidad de información al incluir la estimación de la densidad en ambos lados de la categoría.


```python
# Gráfico de violín para CreditScore por Tipo de Tarjeta
sns.violinplot(x='CreditScore', data=df)

# Mostrar gráfico
plt.title('Distribución de CreditScore')
plt.show()
```

## Heatmap de Correlaciones

El heatmap o mapa de calor es una representación gráfica de una matriz de datos donde los valores se representan mediante variaciones de color. En este caso, podemos usar un heatmap para visualizar la matriz de correlación entre las variables numéricas en nuestro dataset. Esto nos permite identificar rápidamente qué variables están más relacionadas entre sí.

### 1. Heatmap de la Matriz de Correlación

Este gráfico muestra la correlación entre las variables numéricas en el dataset del churn bancario. Los valores cercanos a 1 indican una alta correlación positiva, mientras que los valores cercanos a -1 indican una alta correlación negativa.


```python
# Matriz de correlación
correlation_matrix = df.corr()
#Sólo sirve con columnas numéricas
```

```python
# Seleccionar solo las columnas numéricas
df_numeric = df[['CreditScore', 'Edad', 'Estadia', 'Balance', '#Prod', 'TarjetaCredito', 
                 'Activo', 'Salario', 'Salida', 'Quejas', 'Nivel_satisf', 'Puntos']]
# Calcular la matriz de correlación
correlation_matrix = df_numeric.corr()
correlation_matrix
```

```python
plt.figure(figsize=(10,10))
# Gráfico de calor
sns.heatmap(correlation_matrix, annot=True, cmap='inferno')

# Mostrar gráfico
plt.show()
```

## Pairplot
El pairplot se utiliza para visualizar todas las combinaciones de pares de variables en un DataFrame. Es especialmente útil para explorar relaciones entre múltiples variables numéricas y ver cómo se distribuyen en función de una variable categórica.

Uso: Visualizar relaciones entre varias variables numéricas y categóricas.
Características:
Muestra un gráfico de dispersión (scatter plot) para cada par de variables.
En la diagonal, generalmente muestra histogramas o gráficas de densidad para cada variable.

```python
# Pairplot para visualizar la relación entre las variables numéricas
sns.pairplot(df2_numeric)
# Mostrar gráfico
plt.title('Pairplot de Variables Numéricas')
plt.show()
```

## Jointplot
El jointplot se utiliza para mostrar la relación entre dos variables y también su distribución marginal. Es útil para ver cómo se correlacionan dos variables numéricas y también proporciona histogramas de las distribuciones de cada variable en los ejes.

Uso: Visualizar la relación entre dos variables y sus distribuciones.
Características:
Muestra un gráfico de dispersión entre dos variables.
Incluye histogramas o gráficos de densidad en los márgenes.

```python
# Gráfico conjunto de Salario y Balance
sns.jointplot(data=df, x='Salario', y='Balance', hue='Pais', kind='scatter')
plt.title('Jointplot de Salario y Balance')
plt.show()
```

## Displot
El displot es una función que combina histogramas y gráficos de densidad. Es útil para visualizar la distribución de una sola variable numérica y se puede usar con la opción de hue para mostrar diferentes categorías.

Uso: Visualizar la distribución de una variable.
Características:
Puede mostrar un histograma y/o un gráfico de densidad.
Permite dividir el gráfico por categorías usando col o row.
Tiene la opción de ajustar el número de bins y cambiar la apariencia.

```python
# Gráfico de distribución para Salario
sns.displot(data=df, x='Salario', hue='Genero', kde=True, bins=30)

# Mostrar el gráfico
plt.title('Distribución del Salario por Género')
plt.show()
```
