```python
#Se carga la base de datos

import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/jarsmp37/introds/main/Churn_Bank.csv')
df.head()
```

```python
#Eliminar una fila por el indexado

# Elimina filas con índices 0 y 1
df = df.drop([0, 1])
df
```

```python
# Restablecer el índice
df = df.reset_index(drop=True)
df
```

```python
#Eliminar columna en específico
df2= df.drop(columns=['RowNumber'])
df2
```

```python
#Renombrar una columna
df2 = df.rename(columns={'Surname': 'Apellido'})
df2
```

```python
# Crear un rango de fechas
start_date = '2024-01-01'
end_date = '2024-12-31'
date_range = pd.date_range(start=start_date, end=end_date, periods=12205)

# Convertir las fechas a formato string
date_range_str = date_range.strftime('%Y-%m-%d')
date_range_str 
```

```python
#Anexar una columna llamada date
df['date']=date_range_str 
df.head()
```

```python
df.info()
```

```python
df['date'] = pd.to_datetime(df['date'])
df.info()
```

```python
# Filtrar por una fecha específica
df_filtrado = df[df['date'] == '2024-06-01']
df_filtrado
```

```python
# Filtrar por un rango de fechas
df_rangofecha = df[(df['date'] >= '2024-01-01') & (df['date'] <= '2024-06-30')]
df_rangofecha
```

```python
# Extraer el año
df['año'] = df['date'].dt.year

# Extraer el mes
df['mes'] = df['date'].dt.month

# Extraer el día
df['dia'] = df['date'].dt.day

# Extraer el día de la semana
df['díanombre'] = df['date'].dt.day_name()

df.head()
```

```python
# Calcular la diferencia en días entre dos fechas
df['days_difference'] = (df['date'] - pd.Timestamp('2024-01-01')).dt.days
df.tail()
```

```python
# Diccionario para traducir los nombres de los días de la semana
traduccion = {
    'Monday': 'Lunes',
    'Tuesday': 'Martes',
    'Wednesday': 'Miércoles',
    'Thursday': 'Jueves',
    'Friday': 'Viernes',
    'Saturday': 'Sábado',
    'Sunday': 'Domingo'
}
```

```python
# Reemplazar los nombres de los días en la columna 'dianombre'
df['díanombre'] = df['díanombre'].replace(traduccion)
df
```

```python
#Guardar los resultados en un csv
df.to_csv("Base_limpia.csv", index=False)
```
