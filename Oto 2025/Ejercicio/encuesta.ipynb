```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df=pd.read_csv("ejmp_comida.csv")
df
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df2=df[df["Genero"]=="Femenino"]
df2
```

```python
plt.figure(figsize=(5, 4)) 
df2['Mascota'].value_counts().plot(kind='bar', color='green')
plt.title('Comidas al d√≠a')
plt.show()
```
