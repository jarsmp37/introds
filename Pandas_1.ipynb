Creación de un data frame a partir de un diccionario

```python
import pandas as pd
data={'p':[True, True,False,False],'q':[True,False,True,False]}
df=pd.DataFrame(data)
df
```

Anexar una columna

```python
df['p and q']= [p and q for p, q in zip(df['p'], df['q'])]
df
```

Anexar otra columna sin quitar la primera

```python
df['(p and q) or q']= [not (p and q) or q for p, q in zip(df['p'], df['q'])]
df
```
