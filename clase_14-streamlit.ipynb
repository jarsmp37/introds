1. Layouts en Streamlit

Streamlit permite organizar el contenido en columnas, expansores y secciones laterales (sidebar).
Esto facilita crear interfaces m치s ordenadas e intuitivas.

游빔 Conceptos principales

st.sidebar: permite mostrar widgets en una barra lateral.

st.columns: divide la pantalla en columnas.

st.expander: crea secciones colapsables para organizar informaci칩n adicional.

```python
import streamlit as st

st.title("Ejemplo de Layouts en Streamlit")

# Sidebar
st.sidebar.header("Configuraciones")
nombre = st.sidebar.text_input("Nombre del usuario:")
mostrar_info = st.sidebar.checkbox("Mostrar informaci칩n adicional")

# Columnas
col1, col2 = st.columns(2)
col1.write("Columna 1: puede contener texto, im치genes o gr치ficos.")
col2.write("Columna 2: 칰til para comparaciones o m칠tricas.")

# Expander
with st.expander("Ver detalles"):
    st.write("Este texto est치 dentro de un expander y se puede ocultar.")
    st.image("https://streamlit.io/images/brand/streamlit-logo-secondary-colormark-darktext.png", width=150)

if mostrar_info:
    st.success(f"Bienvenido, {nombre}! Has activado la opci칩n de informaci칩n adicional.")
```

2. Pesta침as (Tabs)

Streamlit permite organizar contenido en pesta침as mediante st.tabs().
Esto es ideal para separar an치lisis, reportes o visualizaciones dentro de la misma app.

```python
import streamlit as st

st.title("Ejemplo de pesta침as en Streamlit")

tab1, tab2, tab3 = st.tabs(["Datos", "Gr치ficas", "Conclusiones"])

with tab1:
    st.write("Aqu칤 puedes mostrar tablas o descripciones de datos.")

with tab2:
    st.line_chart({"Ventas": [100, 150, 200, 180, 220]})

with tab3:
    st.write("Conclusi칩n: las ventas muestran una tendencia creciente con una ligera ca칤da en el cuarto periodo.")
```

3. Filtros din치micos

Streamlit facilita crear filtros para interactuar con los datos sin modificar el c칩digo.
Podemos usar selectores (selectbox, slider, etc.) para actualizar las visualizaciones en tiempo real.

```python
import streamlit as st
import pandas as pd
import numpy as np

st.title("Filtros din치micos con Streamlit")

# Datos de ejemplo
np.random.seed(0)
df = pd.DataFrame({
    "Ciudad": np.random.choice(["Puebla", "CDMX", "Monterrey"], size=30),
    "Ventas": np.random.randint(500, 5000, size=30),
    "Mes": np.random.choice(["Enero", "Febrero", "Marzo", "Abril"], size=30)
})

# Filtros
ciudad = st.selectbox("Selecciona una ciudad:", df["Ciudad"].unique())
mes = st.multiselect("Selecciona uno o varios meses:", df["Mes"].unique(), default=df["Mes"].unique())

# Aplicar filtros
df_filtrado = df[(df["Ciudad"] == ciudad) & (df["Mes"].isin(mes))]

st.write("### Datos filtrados:")
st.dataframe(df_filtrado)

# Gr치fico
st.bar_chart(df_filtrado["Ventas"])
```

5. Mini Dashboard Interactivo

Este ejemplo combina los conceptos anteriores: sidebar, pesta침as, filtros y gr치ficos din치micos.

```python
import streamlit as st
import pandas as pd
import numpy as np
import plotly.express as px

st.set_page_config(page_title="Dashboard de Ventas", layout="wide")

st.title("游늵 Dashboard Interactivo de Ventas")

# Datos simulados
np.random.seed(1)
df = pd.DataFrame({
    "Ciudad": np.random.choice(["Puebla", "CDMX", "Guadalajara", "Monterrey"], 100),
    "Mes": np.random.choice(["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio"], 100),
    "Ventas": np.random.randint(500, 5000, 100)
})

# Sidebar con filtros
st.sidebar.header("Filtros")
ciudad = st.sidebar.selectbox("Selecciona una ciudad:", ["Todas"] + list(df["Ciudad"].unique()))
meses = st.sidebar.multiselect("Selecciona meses:", df["Mes"].unique(), default=df["Mes"].unique())

# Aplicar filtros
df_filtrado = df[df["Mes"].isin(meses)]
if ciudad != "Todas":
    df_filtrado = df_filtrado[df_filtrado["Ciudad"] == ciudad]

# Layout de columnas
col1, col2 = st.columns(2)

with col1:
    st.subheader("Datos filtrados")
    st.dataframe(df_filtrado)

with col2:
    st.subheader("Gr치fico de ventas")
    fig = px.bar(df_filtrado, x="Mes", y="Ventas", color="Ciudad", barmode="group")
    st.plotly_chart(fig, use_container_width=True)

# M칠tricas
st.markdown("---")
st.subheader("Resumen general")

total_ventas = df_filtrado["Ventas"].sum()
promedio = df_filtrado["Ventas"].mean()

col3, col4 = st.columns(2)
col3.metric("Total de ventas", f"${total_ventas:,.0f}")
col4.metric("Promedio de ventas", f"${promedio:,.2f}")
```

Dashboard avanzado con KPIs autom치ticos

En un dashboard real, solemos mostrar indicadores clave de desempe침o (KPIs) como ventas totales, promedio o m치ximo.
Podemos usar st.metric() para mostrarlos con formato profesional.

```python
import streamlit as st
import pandas as pd
import numpy as np
import plotly.express as px

st.set_page_config(page_title="Dashboard Avanzado", layout="wide")

st.title("游늵 Dashboard Avanzado de Ventas")

# Simular datos
np.random.seed(42)
df = pd.DataFrame({
    "Ciudad": np.random.choice(["Puebla", "CDMX", "Guadalajara", "Monterrey"], 200),
    "Mes": np.random.choice(["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio"], 200),
    "Ventas": np.random.randint(500, 10000, 200)
})

# Filtros en sidebar
st.sidebar.header("Filtros")
ciudades = st.sidebar.multiselect("Selecciona ciudades:", df["Ciudad"].unique(), default=df["Ciudad"].unique())
meses = st.sidebar.multiselect("Selecciona meses:", df["Mes"].unique(), default=df["Mes"].unique())

df_filtrado = df[(df["Ciudad"].isin(ciudades)) & (df["Mes"].isin(meses))]

# KPIs
total_ventas = df_filtrado["Ventas"].sum()
promedio_ventas = df_filtrado["Ventas"].mean()
max_ventas = df_filtrado["Ventas"].max()

col1, col2, col3 = st.columns(3)
col1.metric("游눯 Total de Ventas", f"${total_ventas:,.0f}")
col2.metric("游늳 Promedio", f"${promedio_ventas:,.2f}")
col3.metric("游끥 Venta M치xima", f"${max_ventas:,.0f}")

# Gr치fico principal
fig = px.bar(df_filtrado, x="Mes", y="Ventas", color="Ciudad", barmode="group",
             title="Ventas por Ciudad y Mes")
st.plotly_chart(fig, use_container_width=True)

# Gr치fico secundario
fig2 = px.box(df_filtrado, x="Ciudad", y="Ventas", color="Ciudad", title="Distribuci칩n de Ventas por Ciudad")
st.plotly_chart(fig2, use_container_width=True)
```
