## Buscar registros duplicados

Por defecto, df.duplicated() considera una fila como duplicada si todos los valores de la fila coinciden exactamente con otra fila anterior en el DataFrame.

```python
#Con este comando se crea un booleano que muestra renglones duplicados
df.duplicated()
```

Muestra la suma de todas las filas duplicadas

```python
df.duplicated().sum()
```

Puedes especificar un subconjunto de columnas para considerar al determinar duplicados

```python
#Compara solo las columnas 'CustomerId', 'Surname' por lo que no 
#es requisito que sean las 18 columnas
#exactamente iguales, sÃ³lo las columnas 'CustomerId', 'Surname'
df.duplicated(subset=['CustomerId', 'Surname']).sum()
```

# Elimina los registros duplicados

```python
df_sin_dupl=df.drop_duplicates(subset=['CustomerId', 'Surname'])
df_sin_dupl
```

Se cuentan quitar duplicados, cabe aclarar que es sobre la original

```python
df_sin_dupl.duplicated().sum()
```

Vamos a generar un ciclo for que me indique cuantas veces aparece el texto invalid_value en todas las columnas

# Cambio del tipo de valores

A veces nuestros datos que observamos que son numÃ©ricos el df los identifica como caracter, se utiliza el comando info() para ver el Dtype

```python
df2.info()
```

```python
df2.head(3)
```

Se observa que columnas como HasCrCard es numÃ©rico, pero el dtype nos dice que es objeto, para verificar esto se utiliza el comando unique que nos muestra los valores Ãºnicos por columna

```python
df2['HasCrCard'].unique()
```

Se observa que los nÃºmeros estÃ¡n entre comillas, eso indica que son caracteres y no nÃºmeros.

Para cambiar el tipo se ocupa el comando astype( tipo de valor)

int  entero

float decimal

```python
df2['HasCrCard']=df2['HasCrCard'].astype(int)
```

si se verifica en unique de la columna tendremos que ya los considera como valores enteros, al no aparecer con comillas y el dtype object

```python
df2['HasCrCard'].unique()
```

Se visualiza que con info() ya cambio su tipo a int

```python
df2.info()
```

Muchas veces sale error un valor, es recomendable intentar pasar a valor float y despuÃ©s a int, para estos casos

Este proceso se recomienda hacerlo de forma manual, ya que el proceso de automatizaciÃ³n puede generar que algunas columnas que si tienen caracteres las considere como valor entero, convirtiendo a hexadecimal el valor y modificando la estructura de la columna
