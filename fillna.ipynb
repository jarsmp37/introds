```python
#Se carga la base de datos

import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/jarsmp37/introds/main/Churn_Bank.csv')
df.head()
```

```python
lista_col=df.columns
df1=df
for i in lista_col:
    df1=df1[df1[i] != 'invalid_value']
df1
```

```python
for i in lista_col:
    print(f"En la columna {i} los invalid_value son: {df1[df1[i] == 'invalid_value'].shape[0]}")
```
